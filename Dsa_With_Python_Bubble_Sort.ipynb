Bubble sort algorithm

```python
def bubble_sort(arr):
  n=len(arr)
  for i in range(n):
    swap=False
    for j in range(n-i-1):
      if arr[j]>arr[j+1]:
        arr[j],arr[j+1]=arr[j+1],arr[j]
        swap=True
    if not swap:
      break
  return arr
arr=[4,1,5,0]
print(bubble_sort(arr))
```

1   Move Zeroes to the End


Problem: Given an integer array, move all 0s to the end while maintaining the relative order of the non-zero elements. Approach: This uses modified Bubble Sort logic. Instead of bubbling the absolute largest element, you bubble the 0s to the right by swapping them with adjacent non-zero elements.

```python
def move_zero_to_end_using_bubble_sort(arr):
  n=len(arr)
  for i in range(n):
    swap=False
    for j in range(n-i-1):
      if arr[j]==0 and arr[j+1]!=0:
        arr[j],arr[j+1]=arr[j+1],arr[j]
        swap=True
    if not swap:
      break
  return arr
arr=[1, 2, 0, 4, 3, 0, 5, 0]
print(move_zero_to_end_using_bubble_sort(arr))
```

2   Sort an Array in Descending Order


Problem: Modify standard Bubble Sort to sort elements from highest to lowest. Approach: Simply flip the comparison operator. If arr[j] < arr[j + 1], swap them. This forces the smallest elements to "bubble" to the end.

```python
def sort_descanding_arr(arr):
  n=len(arr)
  for i in range(n):
    swap=False
    for j in range(n-i-1):
      if arr[j]<arr[j+1]:
        arr[j],arr[j+1]=arr[j+1],arr[j]
        swap=True
    if not swap:
      break
  return arr
arr=[12,40,15,67,39,90,2,202]
print(sort_descanding_arr(arr))
```

3  Sort an Array of Strings by Length

Problem: Given an array of strings, sort them so that shorter strings appear first. Approach: Use the Bubble Sort iteration, but compare the len(arr[j]) and len(arr[j+1]) instead of the alphabetical characters.

```python
def sort_by_len_using_bubble_sort(arr):
  n=len(arr)
  for i in range(n):
    swap=False
    for j in range(n-i-1):
      if len(arr[j])>len(arr[j+1]):
        arr[j],arr[j+1]=arr[j+1],arr[j]
        swap=True
    if not swap:
      break
  return arr
arr=['100','10','2000','50000','1005','2','100000']
print(sort_by_len_using_bubble_sort(arr))
```

Check if an Array is Sorted in One Pass
Problem: Write a function that returns True if an array is sorted, using the mechanics of Bubble Sort. Approach: Utilize the swapped flag optimization. Run exactly one pass of the inner loop. If no swaps occur, return True. This executes in O(n) time.

```python
def is_sorted(arr):
  n=len(arr)
  swap=False
  for j in range(n-1):
    if arr[j]>arr[j+1]:
      swap=True
      break
  if not swap:
    return True
  else:
    return False
arr=[1,2,3,4,5]
print(is_sorted(arr))
```

5  Sort by Absolute Difference from a Target Value X

Problem: Given an array and a value X, sort the array based on the absolute difference between the elements and X. Approach: During the adjacent element comparison step, evaluate abs(arr[j] - X) > abs(arr[j+1] - X). If true, swap the elements.

```python
def sort_by_absolute_difference_using_bubble_sort(arr,x):
  n=len(arr)
  for i in range(n):
    swap=False
    for j in range(n-i-1):
      if abs(arr[j]-x)>abs(arr[j+1]-x):
        arr[j],arr[j+1]=arr[j+1],arr[j]
        swap=True
    if not swap:
      break
  return arr
arr=[10, 5, 3, 9, 2]
x=7
print(sort_by_absolute_difference_using_bubble_sort(arr,x))
```
